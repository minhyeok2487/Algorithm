```python
N = int(input())

queue_list = list()
for i in range(N):
    queue_list.append(i+1)
    
put_list = list()
for i in range(N):
    if len(queue_list) > 1:
        put_list.append(queue_list.pop(0))
        queue_list.append(queue_list.pop(0))
    else:
        queue_list.append(queue_list.pop(0))
        put_list.append(queue_list.pop(0))

for i in range (0, N ):
    print(put_list[i], end = ' ')
```
