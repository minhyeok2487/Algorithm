## 1. 입력된 숫자를 바꿔서 확인하는 함수

```python
def convert(number):
    convert_number = []
    for i in number:
        if i == '1':
            convert_number.append('c')
        elif i == '2':
            convert_number.append('d')
        elif i == '3':
            convert_number.append('e')
        elif i == '4':
            convert_number.append('f')
        elif i == '5':
            convert_number.append('g')
        elif i == '6':
            convert_number.append('a')
        elif i == '7':
            convert_number.append('b')
        elif i == '8':
            convert_number.append('C')
    main = ''
    for i in convert_number:
        main += i
    return main

def check(convert_number):
    if convert_number == "cdefgabC":
        print('ascending')
    elif convert_number == "Cbagfedc":
        print('descending')
    else:
        print('mixed')

def musical_scales(number):
    number.split(' ')
    convert_number = convert(number)
    check(convert_number)

number = input()
musical_scales(number)
```

## 2. 오름차순 내림차순만을 비교하여 출력하는 함수

```python
def sort_check(number):
    ascending = True
    decending = True
    
    for i in range(1, 8):
        if number[i] > number[i-1]:
            decending = False
        elif number[i] < number[i-1]:
            ascending = False

    if ascending:
        print('ascending')
    elif decending:
        print('descending')
    else:
        print('mixed')

number = list(map(int, input().split(' ')))
sort_check(number)
```
